#Provide an example of a simple Python program that uses both NumPy and Pandas to generate a dataset, create a DataFrame, perform basic operations like adding a new column, sorting, filtering, and calculating statistics?

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(0) 
data = {
    'ID': np.arange(1, 11),  
    'Age': np.random.randint(18, 60, size=10),  
    'Score': np.random.randint(50, 101, size=10)  
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Category'] = np.where(df['Score'] >= 75, 'High', 'Low')
```

```python
df
```

```python
df_sorted = df.sort_values(by='Age')
```

```python
df_sorted
```

```python
df_high_category = df[df['Category'] == 'High']
```

```python
df_high_category
```

```python
age_mean = df['Age'].mean()
score_mean = df['Score'].mean()
age_max = df['Age'].max()
score_min = df['Score'].min()
```

```python
print("\nStatistics:")
print(f"Mean Age: {age_mean:.2f}")
print(f"Mean Score: {score_mean:.2f}")
print(f"Max Age: {age_max}")
print(f"Min Score: {score_min}")
```
